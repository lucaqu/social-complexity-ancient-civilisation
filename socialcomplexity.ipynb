## **Social complexity in ancient civilisations**

### 1. Introduction/Background 

The transition from small-scale human groups to complex, multilevel, and hierarchically structured human societies has long been a scientific puzzle. Major transitions in social complexity over evolutionary time are of extraordinary interest to scholars from widely different fields such as archaeology, anthropology, biology, psychology, sociology, and philosophy. This data set is about the change of social complexity in various ancient human societies across the globe, measured by means of different variables.

A next useful step is to visualise our data, e.g. by plotting a correlation matrix to get a grasp of how our variables relate to one another. Several of our variables measure complexity of certain characteristics in our ancient societies. 

```python
#drop one column
social_drop = social.drop(columns=['MG_corr'])
#quantify correlation matrix
correlationmatrix= social_drop.corr()
#specify characteristics of correlation matrix plot
fig = plt.figure(figsize = (11, 9))
#plot heat map of correlation matrix, specify shape and colour
sns.heatmap(correlationmatrix, square = True,cmap = "PiYG")
#title for plot of correlation matrix
plt.title("Correlation matrix of the social complexity data set", fontweight='bold',fontsize=14)
#display plot of correlation matrix
plt.show()
```

We can also create a different heatmap of our correlation matrix as another way of visualising how different variables are correlated in our data set.

```python
#calculate correlation matrix of our data set
corr_matrix = social.corr()

#create a for loop to mask the redundant half of the symmetrical correlation matrix
mask = []
for i in range(len(corr_matrix.columns)):
    mask_i = []
    for j in range(len(corr_matrix.columns)):
        if i>j:
            mask_i.append(True)
        else: 
            mask_i.append(False)
    mask.append(mask_i)
```

```python
#plot of our heatmap based on the correlation matrix
plt.figure(figsize=(11, 9), facecolor='w', edgecolor='k')
#specify title for heat map of correlation matrix
plt.title("Heat map: correlation matrix of the social complexity data set", fontweight='bold',fontsize=14)
sns.set(font_scale=1.2)
#specify further characteristics of heat map
sns.heatmap(corr_matrix,cmap='coolwarm',
            center = 0, 
            annot=True,
            fmt='.1g',
            mask=mask)
```

Given that we have a data set about how social complexity changes over time, it is valuable to examine what the time periods are for different NGAs. To this end, we plot a timeline.

```python
#group data by time and NGA
socialtime = social.groupby('NGA')['Time']
#create column with year of first data point per NGA, i.e.'beginning'
social['Beginning']= socialtime.transform('min')
#create column with year of last data point per NGA, i.e. 'end'
social['End']= socialtime.transform('max')
```

```python
#plot figure and define figure size
plt.figure(figsize=(15,11))
#create a scatter plot with '-' as a marker to make it look like a timeline
plt.scatter(social['Time'],social['NGA'],marker= '_', color='olive')
#specify label for x-axis
plt.xlabel("Time (Year)",fontsize=12)
#specify label for y-axis
plt.ylabel("NGA",fontsize=12)
#specify figure title
plt.title("Data for NGAs over time", fontweight='bold',fontsize=14)
#slightly change timeline layout
plt.tight_layout()
```

We see that we have NGAs over vastly different time periods, and therefore it makes sense to then visualise how often they were sampled. 

```python
#specify figure size
plt.figure(figsize=(12,10))
#specify characteristics of count plot, such as the category to be counted and colour
sns.countplot(social['NGA'], color = 'olive', order = social['NGA'].value_counts().index)
#specify figure title
plt.title('Amount of data per NGA',fontweight='bold',fontsize=14)
#specify label for x-axis
plt.xlabel('NGA',fontsize=12)
#specify label for y-axis
plt.ylabel('Count',fontsize=12)
#rotate label for categories (i.e. NGAs) on x-axis
plt.xticks(rotation= 80)
#display plot
plt.show()
```

An interesting question about this data set is how an overal measure of social complexity (captured by SPC1, a Principal Component created from several variables in the data set) changed over time in ancient societies across the globe. We therefore visualise the cultural evolutionary change of SPC1 over time for NGAs from different continents. 

```python
#assign NGAs to their corresponding continent (Europe, Africa, South America, South-East Asia)
#NGAs of Europe
social_europe = social[(social['NGA']=="Iceland") | (social['NGA']=="Paris Basin") | (social['NGA']=="Latium") | (social['NGA']=="Konya Plain")]
#NGAs of Africa
social_africa = social[(social['NGA']=="Upper Egypt") | (social['NGA']=="Ghanaian Coast") | (social['NGA']=="Niger Inland Delta")]
#NGAs of South-America
social_southamerica = social[(social['NGA']=="Lowland Andes") | (social['NGA']=="North Colombia") | (social['NGA']=="Cuzco")]
#NGAs of South-East Asia
social_southeastasia = social[(social['NGA']=="Cambodian Basin") | (social['NGA']=="Kapuasi Basin") | (social['NGA']=="Central Java") | (social['NGA']=="Oro PNG")]
```

```python
#plot of the change in SPC1 of NGAs in Africa 
#specify figure title  
plt.title('SPC1 of NGAs in Africa over time',fontweight='bold', fontsize = 14)
#specify label for x-axis
plt.xlabel('Time (Year)', fontsize = 12)
#specify label for y-axis
plt.ylabel('Social Complexity (SPC1)', fontsize = 12)
#specify y-limit
plt.ylim(0,1)
#plot SPC1 for NGAs
plt.plot('Time','SPC1',data=social_africa[social_africa['NGA']=="Upper Egypt"], color = 'darkkhaki')
plt.plot('Time','SPC1',data=social_africa[social_africa['NGA']=="Ghanaian Coast"], color = 'indianred')
plt.plot('Time', 'SPC1',data=social_africa[social_africa['NGA']=="Niger Inland Delta"], color = 'lightseagreen')
#add figure legend
plt.legend(['Upper Egypt','Ghanaian Coast','Niger Inland Delta'])
#display figure
plt.show()

#plot of the change in SPC1 of NGAs in Europe
#specify figure title
plt.title('SPC1 of NGAs in Europe over time',fontweight='bold', fontsize = 14)
#specify label for x-axis
plt.xlabel('Time (Year)', fontsize = 12)
#specify label for y-axis
plt.ylabel('Social Complexity (SPC1)', fontsize = 12)
#specify y-limit
plt.ylim(0,1)
#plot SPC1 for NGAs
plt.plot('Time','SPC1',data=social_europe[social_europe['NGA']=="Iceland"], color = 'darkkhaki')
plt.plot('Time','SPC1',data=social_europe[social_europe['NGA']=="Paris Basin"], color = 'indianred')
plt.plot('Time', 'SPC1',data=social_europe[social_europe['NGA']=="Latium"], color = 'lightseagreen')
plt.plot('Time', 'SPC1',data=social_europe[social_europe['NGA']=="Konya Plain"], color = 'fuchsia')
#add figure legend
plt.legend(['Iceland','Paris Basin','Latium', 'Konya Plain'])
#display figure
plt.show()

#plot of the change in SPC1 of NGAs in South America
#specify figure title
plt.title('SPC1 of NGAs in South America over time',fontweight='bold', fontsize = 14)
#specify label for x-axis
plt.xlabel('Time (Year)', fontsize = 12)
#specify label y-axis 
plt.ylabel('Social Complexity (SPC1)', fontsize = 12)
#specify y-limit
plt.ylim(0,1)
#plot SPC1 for NGAs
plt.plot('Time','SPC1',data= social_southamerica[social_southamerica['NGA']=="Lowland Andes"], color = 'darkkhaki')
plt.plot('Time','SPC1',data=social_southamerica[social_southamerica['NGA']=="North Colombia"], color = 'indianred')
plt.plot('Time', 'SPC1',data= social_southamerica[social_southamerica['NGA']=="Cuzco"], color = 'lightseagreen')
#add figure legend
plt.legend(['Lowland Andes','North Colombia','Cuzco'])
#display figure
plt.show()

#plot of the change in SPC1 of NGAs in South-East Asia
#specify figure title
plt.title('SPC1 of NGAs in South-East Asia over time',fontweight='bold', fontsize = 14)
#specify label for x-axis
plt.xlabel('Time (Year)', fontsize = 12)
#specify label for y-axis
plt.ylabel('Social Complexity (SPC1)', fontsize = 12)
#specify y-limit
plt.ylim(0,1)
#plot SPC1 for NGAs
plt.plot('Time','SPC1',data= social_southeastasia[social_southeastasia['NGA']=="Cambodian Basin"], color = 'darkkhaki')
plt.plot('Time','SPC1',data=social_southeastasia[social_southeastasia['NGA']=="Kapuasi Basin"], color = 'indianred')
plt.plot('Time', 'SPC1',data=social_southeastasia[social_southeastasia['NGA']=="Central Java"], color = 'lightseagreen')
plt.plot('Time', 'SPC1',data=social_southeastasia[social_southeastasia['NGA']=="Oro PNG"], color = 'fuchsia')
#add figure legend
plt.legend(['Cambodian Basin','Kapuasi Basin','Central Java', 'Oro PNG'])
#display figure
plt.show()
```

## 3. Summary/Conclusion

We have seen that ancient civilisations across the globe differed quite remarkably in social complexity, but there were no visible differences between continents. Many societies underwent a marked increase in social complexity over time, and it seemed that older NGAs reached higher values overall. Some of the variables characterising social complexity were highly positively correlated, and therefore it was sensible to use SPC1 as a proxy for social complexity. This also suggests that several measures of the complexity of human societies might be closely linked and evolve dependently on each other.
