The social complexity data set uses an overal measure of social complexity (captured by SPC1, a Principal Component created from several variables in the data set). Here, we replicate the Principal Components Analysis based on nine variables, using own code. To this end, we first import several libraries necessary to conduct our PCA.

```python
#import libraries for PCA
from sklearn import decomposition
from sklearn import datasets
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
import numpy as np
from pandas import DataFrame
```

```python
#create a new data frame by removing variables that are not relevant - only certain variables are of interest for the PCA
social1 = social.drop(columns=['MG_corr', 'NGA', 'PolID', 'Time', 'SPC1', 'Beginning', 'End'])
social1 
```

```python
#scale variables in the data set which are at different levels - an important requirement for PCA 
X = StandardScaler().fit_transform(social1)
X
```

```python
#perform our PCA, specify number of principal components
pca = decomposition.PCA(n_components=2)
X = pca.fit_transform(X)
X
```

```python
#create PCA data frame with same number of rows as in original data frame, with PC1 and PC2 as columns
pcadf = DataFrame(data = X
             , columns = ['principal component 1', 'principal component 2'])
pcadf
```

```python
#calculate how much variance in our data set is explained by PC1 and PC2
pca.explained_variance_ratio_
```

PC1 in our PC seems to explain 82.47 percent of the variation in our data set, which is a remarkable value. However, it seems that we do not exactly replicate with our own code what has been found in the original paper (75 percent).

We are also interested in the loadings of the PCA, i.e. how variables in the data set load onto the principal components. 

```python
#obtain loadings of our PCA to evaluate how the variables in the data set load onto the principal components
loadings = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2'])
loadings
```

Strikingly, we can see that all nine variables seem to load onto PC1 - which explains the vast majority of variation - quite similarly. This is not surprising given the high level of correlation between the variables. The different measures of social complexity do seem interdependent, which means that it is justifiable to use SPC1 as an overall proxy for social complexity.
